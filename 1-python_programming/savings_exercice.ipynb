How much do you get for your savings ?
1. let's create a function that will allow us to know how much money a user will have in total after a certain number of years at a certain rate of interest

2. The user should be able to call the function which will then ask for : - the total amount he wishes to place - The number of years he wants to invest his money for - The rate of interest to which he is entitled

3. Pay attention to the errors that the user may enter :

- if he enters a string of characters rather than a number, an error should be displayed
- if the user enters negative numbers, an error should also be output
- finally, it is very likely that the user will write the interest rate as a percentage (e.g. 10%). Raise an error if this is the case, to prevent him from inputing a percentage.
Finally, the program should only show number with a maximum of 2 digits after the decimal point.

Example of output :

1000.0

10.0

0.1

The total amount of money you will have after you deposit 1000.00 at the end of 10.0 years will be 2593.74 €

```python
def ask_amount():
    amount = input("Quel montant souhaitez-vous placer ? ")
    try:
        float(amount)
        if float(amount) <= 0:
            print("\n****** Merci d'entrer un montant supérieur à 0")
            amount = ask_amount()
        else:
            return amount
    except ValueError:
        print("\n****** Merci d'entrer un montant en chiffres")
        amount = ask_amount()
    return amount
```

```python
def ask_years():
    years = input("\n Pendant combien d'années souhaitez-vous placer cet argent ? ")
    try:
        int(years)
        if int(years) <= 0:
            print("\n****** Merci d'entrer un nombre d'années supérieur à 0")
            amount = ask_years()
        else:
            return years
    except ValueError:
        print("\n****** Merci d'entrer un nombre d'années en chiffres.")
        years = ask_years()
    return years
```

```python
def ask_roi():
    roi = input("\n A quel taux ? ")
    try:
        float(roi)
        if float(roi) > 1 :
            raise ValueError()
        else:
            if float(roi) <= 0:
              print("\n****** Merci d'entrer un taux supérieur à 0")
              roi = ask_roi()
            else:
              return roi
    except ValueError:
      print("Ceci n'est pas un taux décimal compris entre 0 et 1")
    else:
      return roi
```

```python
def calculate_saving(amount, years, roi):
    total = float(amount) * int(years) * float(roi)
    print(f"\n------------------\nSi vous placez {amount}€ durant {years} ans à un taux de {float(roi)*100}%")
    print(f"Vous gagnerez : {format(total, '.2f')} €")
    print(f"La somme totale sur votre compte sera donc de {format(float(amount) + float(total), '.2f')} €")
```

```python
# main function
def launch_evaluation():
    try:
      amount = ask_amount()
      years = ask_years()
      roi = ask_roi()
      calculate_saving(amount, years, roi)
    except:
      print("Interruption du programme")
```

```python
# lancement du script de calcul d'économies :
launch_evaluation()
```
